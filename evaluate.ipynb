```python
%load_ext dotenv
%dotenv

from tqdm import tqdm
from datasets import Dataset
import pandas as pd
from ragas import evaluate
from langchain.embeddings import HuggingFaceEmbeddings


from ragas.metrics import (
    faithfulness,
    context_relevancy
)

file = 'v2/testset_TAIDE-LX-7B-Chat_RAG.csv'
df = pd.read_csv(file)

results = []
bz = 25
batch_size = len(df)//25+1
for idx in tqdm(range(batch_size)):
    df_dict = {
    "question": df['Question'][idx*bz:(idx+1)*bz],
    "answer": df['Answer_TAIDE-LX-7B-Chat_RAG'][idx*bz:(idx+1)*bz],
    "contexts": [[d] for d in df['Retrieve_Expand_Content'][idx*bz:(idx+1)*bz]],
    "ground_truths": [[d] for d in df['Answer'][idx*bz:(idx+1)*bz]],
    }

    data = Dataset.from_dict(df_dict)
    
    try:
        result = evaluate(
            dataset = data, 
            metrics=[
                    faithfulness,
                    context_relevancy
                    ],
        )
        results.append(result)
    except:
        results.append(0)
        print('*ERROR*\n', df_dict )
    # break
```
